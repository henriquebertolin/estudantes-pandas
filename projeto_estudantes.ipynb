```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
dados = pd.read_csv(url, sep=',')
dados.head()
```

1) Criando uma coluna chamada "Pontos_extras" que contém os pontos extras de cada aluno, ou seja, 40% da nota atual deles.

```python
# tratando valores nulos
dados = dados.fillna(0)
dados.head()
```

```python
dados['Pontos_extras'] = dados['Notas'] * 0.40
dados.head()
```

2) Criando coluna chamada "Notas_finais" que possui as notas de cada aluno somada com os pontos extras.

```python
dados['Nota_final'] = dados['Notas'] + dados['Pontos_extras']
dados.head()
```

3) Criando função lambda que valida se o aluno alcançou a nota máxima

```python
dados['Nota_final'] = dados['Nota_final'].apply(lambda x: 10.0 if x > 10 else x)
dados.head()
```

4) Refatorando valores da coluna "Aprovado" com função que indica se, com a nota extra, o aluno foi aprovado ou não

```python
dados['Aprovado'] = dados['Nota_final'].apply(lambda x: x >= 6)
dados.head()
```

5) Criando uma seleção com os alunos que foram aprovados graças à nota extra

```python
selecao = (dados['Notas'] < 6) & (dados['Nota_final'] > 6)
dados[selecao]
```
